Funciones de agregacion

```python
Ventas = np.random.randint(1, 10, size=20).reshape(4, 5)#Cada fila es 1 semana, y cada columna es un dia de la semana
print(Ventas)
```

```python
print(Ventas)
print(Ventas.max())
print(Ventas.min())
print(Ventas.prod())#Multiplica todos los elementos del array.
print(Ventas.cumsum(axis=1))#Va sumando y poniendo los elementos de las filas (axis=0 son para las columnas).
```

```python
print(Ventas.argmax(), Ventas.argmin())
```

###Producto escalar

```python
V = np.array([1, 2, 3, 4])
W = np.array([5, 6, 7, 8])
print("V@W = ", V@W, ", [1, 2, 3, 4].[5, 6, 7, 8] = 1.5+2.6+7.7+4.8 = ", V@W)
print(V@W, " = ", V.dot(W))
```

```python
A = np.array([[2, 5, 3], [7, 1, 4]])
B = np.array([[3, 2], [9, 6], [5, 8]])
print( "A.B = \n", A.dot(B))
```

```python
a = np.arange(49).reshape(7, 7)
print(a)
print(a.T, " = \n", a.transpose())
```
