```python
x = np.array([0, 1, 2, 3, 4, 5, 6, 7])
plt.plot(x, "o-r", label="Lineal")
plt.plot(x**2, "s:g", label="Cuadratica")
plt.plot(x*3, "D:g", label="Triple")
plt.title("3 funciones")
plt.xlabel("Eje x")
plt.ylabel("Eje Y")
plt.grid(color="c", linestyle="-", linewidth=1)
plt.minorticks_on()
plt.legend()
```

```python
Grupos = ["Ninos", "Jovenes", "Adultos", "Ancianos"]
Hombres = [28, 45, 34, 88]
Mujeres = [10, 72, 37, 65]

Indice = np.arange(len(Grupos))

## Se crean las primeras barras.
plt.barh(Indice, Hombres, label="Hombres")

##Se crean las segundas barras y se aplican sobre las primeras.
plt.barh(Indice, Mujeres, label="Mujeres", bottom=Hombres)

plt.xticks(Indice, Grupos)
plt.ylabel("Participantes")
plt.xlabel("Grupos")
plt.title("Participantes por genero")
plt.legend()

plt.show()
```

```python
GruposEtarios = ["Ninos", "Adolecentes", "Jovenes", "Adultos", "Ancianos"]
Hombres = [30, 15, 20, 30, 10]
Mujeres = [35, 10, 25, 35, 15]
Indice = np.arange(len(GruposEtarios))

plt.bar(GruposEtarios, Hombres, label="Hombres.")
plt.bar(GruposEtarios, Mujeres, label="Mujeres.", bottom=Hombres)

plt.title("Participantes por grupo etario y genero.")
plt.xlabel("Grupos etarios.")
plt.ylabel("Participantes.")

plt.legend()
plt.show()
```
