# GroupBy

```python
DataTitanic.groupby("Sex").count() #Arma una matriz "agrupando" los datos de la columna seleccionada y aplicando la funcion dicha.
```

```python
print(DataTitanic.groupby("Sex").mean()["Age"]) #Promedio de edad en en los sexos.
print(DataTitanic[["Sex", "Age"]].groupby("Sex").mean()) #Mismo resultado.
print(DataTitanic.groupby("Sex")["Age"].mean()) #Mismo resultado.
```

```python
def categoria(x):
    m = x.mean()
    return True if m>29 else False

DataTitanic.groupby("Sex")["Age"].aggregate(["max", "min", "std", "count", categoria])
```

```python
DataTitanic[["Pclass", "Sex", "Age"]].groupby(["Pclass", "Sex"]).aggregate(["count", "mean", "max", "min"])
```

# Missing values
### Tres tipos:
        -type(np.nan) #float
        -type(np.inf) #float
        -type(None) #type None

```python
Df2 = pd.DataFrame({"Fiat":[100, 200, 300, 150, 300], "Renault":[300, 240, np.nan, 300, 400], "Volkswagen":[400, 100, 200, 250, 400]})
Df2
```

```python
print(Df2["Fiat"] + Df2["Renault"])
print(Df2["Fiat"].add(Df2["Renault"], fill_value=0)) #Todos los NaN pasan a valer cero.
```

```python
print(Df2.isna())
print(Df2.notna())
```

```python
Df2.isna().sum()
```

```python
print(DataTitanic["Age"].isna().sum())
print(DataTitanic[["Age", "Sex"]].fillna(DataTitanic["Age"].mean())) #Los que tienen NaN en la edad, se los llena con el promedio global.
```

```python
print(Df2.plot())
print(Df2.interpolate().plot()) #El interpolate() pone un dato proedio en e lugar de NaN para poder graficar sin cortes.
```

```python
Df2.interpolate(method="quadratic").plot()
```
